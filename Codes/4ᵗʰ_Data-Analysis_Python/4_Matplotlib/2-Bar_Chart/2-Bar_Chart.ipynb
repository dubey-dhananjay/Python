```python
import matplotlib.pyplot as plt
import pandas as pd
```

## Plotting Bar Chart 

```python
df = pd.read_csv("MelaSales.csv", delimiter = ",")

df.plot(kind="bar", x="Day", title="Mela Sales Report")
plt.ylabel("Sales in Rs")
```

<b>Customising Bar Chart</b>

```python
df = pd.read_csv("MelaSales.csv", delimiter = ",")

df.plot(kind="bar", x="Day", title="Mela Sales Report", color=["red", "yellow", "purple"], edgecolor="Green", linewidth=2, linestyle="--")
plt.ylabel("Sales in Rs")
plt.show()
```
